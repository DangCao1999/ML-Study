```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.cluster import KMeans
```

# Prepare data

```python
X = 100*np.random.random((1000,2))
print(X)
```

```python
print(X[:,0])
```

# Train

```python
kmeans = KMeans(n_clusters =5, random_state=100)
kmeans.fit(X)
```

```python
print("Center cluster:")
print(kmeans.cluster_centers_)
```

```python
print("Label pre:")
pred_label = kmeans.predict(X)
print(pred_label.shape)
```

# Visualize

```python
X.shape[1]
```

```python

fig = plt.figure(figsize=(10,10))
for i in range(X.shape[0]):
    if pred_label[i] == 0:
        color = "green"
    elif pred_label[i] == 1:
        color = "blue"
    elif pred_label[i] == 2:
        color = "yellow"
    elif pred_label[i] == 3:
        color = "cyan"
    else:
        color = "brown"
    plt.scatter(X[i,0], X[i,1], color=color, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color="red", marker="h", alpha=1, s=200)
plt.show()
```

```python
learnLabels = ['green', 'blue', 'yellow','cyan' 'brown',]
```

```python
X_test = [[50,50]]
pred_X_test = kmeans.predict(X_test)[0]
print("Cluster {}: '{}'".format(pred_X_test, learnLabels[pred_X_test]))
```
