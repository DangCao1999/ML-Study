```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.linear_model import LinearRegression 
```

```python
c = np.array([[-40], [-10], [0], [8], [15], [22], [38], [40], [47], [52], [59], [64], [68], [73], [76], [80], [86], [99], [109] ], dtype= float)
f = np.array([[-40, 14, 32, 46, 59, 72, 100, 104, 116.6, 125.6, 138.2,
                147.2, 154.4, 163.4, 168.8, 176, 186.8, 210.2, 228.2 ]], dtype = float).T
```

# Câu a

```python
# Pt đương thẳng
# y = 1.8x + 32
# Biểu diễn đường thằng với dự liệu train

plt.plot(c,f, color='red')
plt.scatter(c,f, color='blue')
plt.show()
```

```python
lr = LinearRegression()
lr.fit(c,f)
b0 = lr.intercept_
b1 = lr.coef_
print(b0)
print(b1)
```

```python
#Phương trình đường thẳng là

print('y =', b1[0][0] , 'x +', b0[0])
```

```python
# dự đoán độ C là bao nhiêu độ F
x_pre = np.array([[-20, -39, 57.8, 89.5, 222.1111111]]).T
y_pre = lr.predict(x_pre)
print(y_pre)
```

```python
#Biểu diễn đường thằng với dự liệu train + dữ liệu dự báo

figure(figsize=(10, 5), dpi=160)
plt.scatter(x_pre, y_pre, color="green", zorder=3)
plt.plot(x_pre,y_pre, color='red', zorder=1)
plt.scatter(c,f, color='blue', zorder=2)
plt.title("Bieu do bieu dien do C va do F du lieu train va du lieu pre")
plt.xlabel('Do C')
plt.ylabel('Do F')
plt.show()
```
