```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x1 = np.array([73.5, 75.0, 76.5, 79.0, 81.5, 82.5, 84.0, 85.0, 86.5, 87.5, 89.0, 90.0, 91.5]) #dientich
x2 = np.array([20,18,17,16,15,14,12,10,8,7,5,2,1]) #kc
y = np.array([1.49,1.50,1.51, 1.54, 1.58, 1.58, 1.60, 1.62,1.63,1.64, 1.66, 1.67, 1.68]) #gia
print(x1.shape)
print(x2.shape)
print(y.shape)
```

# Câu a:

```python
plt.scatter(y, x1)
plt.title("Bieu do gia theo dien tich")
plt.xlabel("Gia (ty dong)")
plt.ylabel("Dien tich (km2)")
plt.show()
```

# Câu b:

```python
plt.plot(y, x2)
plt.title("Bieu do gia theo khoang cach den trung tam")
plt.xlabel("Gia (ty dong)")
plt.ylabel("Khoang cach den thanh pho")
plt.show()
```

# Câu c:

```python
fig0=plt.figure(figsize=(8,3), dpi=80)

axes1 = fig0.add_axes([0,0,0.3,1])
axes2 = fig0.add_axes([0.5,0,0.3,1])

axes1.scatter(x1, y, color='red')
axes2.plot(x2, y, color='blue', linestyle='--')

axes1.set_xlabel("Dien tich (m2)")
axes1.set_ylabel("Gia (ty dong)")
axes1.set_title("Gia can ho theo dien tich")

axes2.set_xlabel("Khoang cach den TT (km)")
axes2.set_ylabel("Gia (Ty dong)")
axes2.set_title("Gia can ho theo khoang cach toi TT")

plt.show()
```
