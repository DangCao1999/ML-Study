# Câu a:

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,20,(1,10))
print(x.T)
print(x.shape)
```

```python
y = np.zeros(10, dtype=int)
for i in range(0,10):
    value = 2*x[0][i]
    y[i] = value
print(y)
```

```python
plt.plot(x.T,y)
plt.show()
```

# Câu b:

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```
