# Vậy chúng ta cần biết linear regression là gì?
    Linear gression là một thuật toán hồi quy tuyến tính, nó cho phép chúng ta tìm ra một được đường thằng (mối liên hệ) chung dựa trên một tập dữ liệu x,y cho trước
    
    Vì nó linear nên là nó sẽ một đường thẳng => chúng ta có được một công thức tổng quát: y = ax + b.
    
    Mối liên hệ chung ở định nghĩa chúng ta có thể thấy nó chính là hệ số a và b trong công thức tổng quát.

## Vậy overfitting là gì?
    Trong máy học có 2 vấn đề khi training 1 model đó là overfitting và underfiting
    Đối với overfitting đó là mô hình sau khi được train thì cho ra được kết quả với tập validation rất tốt nhưng khi đến đoạn chúng ta dự đoán kết quả với test set thì lại nhận được kết quả rất tệ, các bạn có thể hiểu nôm na là học vẹt ấy.
    Chúng ta coi model cần train ra giống như là một em bé vậy khi chúng ta dạy những bài toán cho em bé thì em bé học và làm bài kiếm tra rất tốt nhưng đến khi gặp một bài toán phức tạp hơn tương tự thì em bé không biết cách giải và đưa ra kq ko chính xác.

# Vậy mô hình Lasso Regression là gì? và có liên quan gì đến những điều bên trên?

    Đối với mô hình ridge regression chúng ta thấy rằng nó được cải tiến nhờ vào **thêm** và **giữ lại** những chỉ số có thể tác động vào mô hình với những tỉ lệ tương ứng trong quá trình tính RSS.

# Let's Demo

```python
# import 1 vài thư viện hỗ trợ chúng ta tiền xử lý data nào
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import Lasso
from sklearn import linear_model
from sklearn.model_selection import learning_curve
```

```python
# đọc dataset thử lên xem xét chút nào
# data_read = pd.read_csv('./data/kc_house_data.csv')
data_read = pd.read_csv('./data/Melbourne_housing_FULL.csv')
data_read.head()
```

```python
data_read.nunique()
```

```python
data_read.shape
```

```python
#cols_to_use = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat',  'long', 'sqft_living15', 'sqft_lot15' ]
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'SellerG', 'Distance',
               'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'Regionname', 'BuildingArea',
               'CouncilArea', 'Propertycount']
data_read = data_read[cols_to_use]
data_read.shape
```

```python
data_read.isna().sum()
```

```python
cols_to_fill_zero = ['Car', 'Propertycount', 'Distance', 'Bedroom2', 'Bathroom']
data_read[cols_to_fill_zero] = data_read[cols_to_fill_zero].fillna(0)
data_read.isna().sum()
```

```python
data_read['BuildingArea'] = data_read['BuildingArea'].fillna(data_read.BuildingArea.mean())
data_read['Landsize'] = data_read['Landsize'].fillna(data_read.BuildingArea.mean())
data_read.isna().sum()
```

```python
data_read.dropna(inplace=True)
data_read.isna().sum()
```

```python
data_read = pd.get_dummies(data_read, drop_first=True)
data_read.head()
```

```python
X = data_read.drop('Price', axis=1)
y = data_read['Price']
#print(X)
#print(y)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state=2)
print(train_X.shape)
```

```python
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
train_sizes, train_scores, test_scores = learning_curve(LinearRegression(), train_X, train_y)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
#
# Plot the learning curve
#
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation Accuracy')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.title('Learning Curve')
plt.xlabel('Training Data Size')
plt.ylabel('Model accuracy')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
lasso_reg = linear_model.Lasso(alpha=50, max_iter =1000, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

```python
print(lasso_reg.intercept_)
print(lasso_reg.coef_.size)
```

```python
train_sizes, train_scores, test_scores = learning_curve(Lasso(alpha=50, max_iter =1000, tol=0.1), train_X, train_y, n_jobs=4)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
#
# Plot the learning curve
#
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation Accuracy')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.title('Learning Curve')
plt.xlabel('Training Data Size')
plt.ylabel('Model accuracy')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
